```python
import requests
import re
import json
from urllib.parse import quote
from loa_everyday_init import HEADERS

# HTML íƒœê·¸ ì œê±° í•¨ìˆ˜ (ì „ì—­ ì‚¬ìš© ê°€ëŠ¥)
def strip_html(text):
    return re.sub(r"<[^>]+>", "", text or "")

def get_character_info(name):
    encoded_name = quote(name)
    url = f"https://developer-lostark.game.onstove.com/armories/characters/{encoded_name}"
    res = requests.get(url, headers=HEADERS)

    if res.status_code != 200:
        return "âŒ API ìš”ì²­ ì‹¤íŒ¨ (ë¹„ê³µê°œ ë˜ëŠ” ì¡´ì¬í•˜ì§€ ì•ŠìŒ)"

    try:
        data = res.json()
    except Exception as e:
        return f"âŒ ì‘ë‹µ íŒŒì‹± ì˜¤ë¥˜: {e}"

    # ê¸°ë³¸ ì •ë³´
    profile = data.get("ArmoryProfile", {})
    character_name = profile.get("CharacterName", name)
    class_name = profile.get("CharacterClassName", "?")
    combat_level = profile.get("CharacterLevel", "?")
    ilvl = data.get("ItemMaxLevel") or profile.get("ItemAvgLevel", "?")
    server = profile.get("ServerName", "?")

    # ì¥ë¹„ ì •ë³´
    armory_eq = data.get("ArmoryEquipment")
    if isinstance(armory_eq, dict) and "Equipment" in armory_eq:
        equipment = armory_eq["Equipment"]
    elif isinstance(armory_eq, list):
        equipment = armory_eq
    else:
        return "âš ï¸ ì¥ë¹„ ë°ì´í„°ë¥¼ ë¶ˆëŸ¬ì˜¬ ìˆ˜ ì—†ìŠµë‹ˆë‹¤."

    # ë¬´ê¸° ë° ë°©ì–´êµ¬
    weapon = next((e for e in equipment if "ë¬´ê¸°" in e.get("Type", "")), None)
    armors = [e for e in equipment if "ë¬´ê¸°" not in e.get("Type", "")]

    def extract_reinforce_info(item):
        """+ì¬ë ¨ ë‹¨ê³„ + (ìƒê¸‰ Në‹¨ê³„) + [ì—ìŠ¤ë”] í‘œì‹œ"""
        name = item.get("Name", "")
        tooltip_raw = item.get("Tooltip", "")
        grade = item.get("Grade", "")

        # ì¬ë ¨ ë‹¨ê³„: +ìˆ«ì
        reinforce_level = "?"
        m = re.search(r"\+(\d+)", name)
        if m:
            reinforce_level = f"+{m.group(1)}"

        # ìƒê¸‰ ì¬ë ¨ ë‹¨ê³„
        supreme_text = ""
        try:
            tooltip_json = json.loads(tooltip_raw)
            el5 = tooltip_json.get("Element_005", {})
            value_field = el5.get("value", "") if isinstance(el5, dict) else el5
            clean_text = strip_html(value_field)
            m2 = re.search(r"(\d+)ë‹¨ê³„", clean_text)
            if m2:
                supreme_text = f"({m2.group(1)})"
        except Exception:
            pass

        # ì—ìŠ¤ë” ë¬´ê¸° ì²˜ë¦¬
        prefix = "ì—ìŠ¤ë” " if grade == "ì—ìŠ¤ë”" else ""

        return f"{prefix}{reinforce_level} {supreme_text}".strip()

    weapon_str = extract_reinforce_info(weapon) if weapon else "ë¬´ê¸° ì—†ìŒ"
    armor_lvls = [extract_reinforce_info(a) for a in armors[:5]]

    # ë³´ì„
    armory_gem = data.get("ArmoryGem")
    gems = []
    if isinstance(armory_gem, dict):
        raw_gems = armory_gem.get("Gems")
        if isinstance(raw_gems, list):
            gems = raw_gems

    gem_levels = {"ê²í™”": [], "ì‘ì—´": []}
    for g in gems:
        name = strip_html(g.get("Name", ""))
        level = g.get("Level", "?")
        if "ê²©" in name or "í”¼í•´" in name or "ê²í™”" in name:
            gem_levels["ê²í™”"].append(str(level))
        elif "ì¬ì‚¬ìš©" in name or "ì‘ì—´" in name:
            gem_levels["ì‘ì—´"].append(str(level))

    # ê°ì¸
    engraving_list = []
    engraving_data = data.get("ArmoryEngraving")
    if isinstance(engraving_data, dict):
        effects = engraving_data.get("Effects")
        if not effects or not isinstance(effects, list):
            effects = engraving_data.get("ArkPassiveEffects")
        if isinstance(effects, list):
            for e in effects:
                name = e.get("Name")
                if name:
                    clean_name = strip_html(name).strip()
                    if clean_name:
                        engraving_list.append(clean_name)
    if not engraving_list:
        engraving_list = ["ì—†ìŒ"]

    # ì•„í¬ íŒ¨ì‹œë¸Œ (ArkPassive â†’ Points ëŒ€ì‘)
    arc_points = {"ì§„í™”": 0, "ê¹¨ë‹¬ìŒ": 0, "ë„ì•½": 0}
    ark_passive = data.get("ArkPassive")
    if isinstance(ark_passive, dict):
        points = ark_passive.get("Points")
        if isinstance(points, list):
            for p in points:
                name = strip_html(p.get("Name", "")).strip()
                value = int(p.get("Value", 0))
                if name in arc_points:
                    arc_points[name] = value

        # ìµœì¢… ì¶œë ¥ ë¬¸ìì—´ ìƒì„±
    output = f"""ğŸ§ {character_name} ({class_name})
ì„œë²„: {server}
ì „íˆ¬ Lv.{combat_level} / ì•„ì´í…œ Lv.{ilvl}

ğŸ—¡ï¸ ë¬´ê¸°: {weapon_str}
ğŸ›¡ï¸ ë°©ì–´êµ¬: {' / '.join(armor_lvls)}

ğŸ’ ë³´ì„
- ê²í™”: {' / '.join(gem_levels['ê²í™”']) or 'ì—†ìŒ'}
- ì‘ì—´: {' / '.join(gem_levels['ì‘ì—´']) or 'ì—†ìŒ'}

ğŸ”® ê°ì¸: {', '.join(engraving_list)}

ğŸŒŸ ì•„í¬ íŒ¨ì‹œë¸Œ
ì§„í™”: {arc_points['ì§„í™”']}pt / ê¹¨ë‹¬ìŒ: {arc_points['ê¹¨ë‹¬ìŒ']}pt / ë„ì•½: {arc_points['ë„ì•½']}pt
"""

    # ğŸ ì´ìŠ¤í„°ì—ê·¸ ì¶”ê°€
    easter_eggs = {
        "ì´í•¼": "ê¹ì¹˜ì§€ë§ˆë¼ From ì´í•¼",
        "ë§¤ì¼ì¢‹ì€ë‚ ": "ğŸŒ ì˜¤ëŠ˜ë„ ì¢‹ì€ í•˜ë£¨ ë˜ì„¸ìš”! - From ë§¤ì¼ì¢‹ì€ë‚ ",
        "ëŸ¬ë´‡": "180.72.17 êµµ (ë¶€ì‹¬ìˆìŒ) \nì²«ì‚¬ë‘ ì´ë¯¸ì§€ / ì‹¤>>>ì‚¬ \nì²­ì´ˆ+ê±´ì‹¤+í‡´íë¯¸+ê³ ê¸‰ë¯¸ \nêµ°ì‚´x, ì²´ìœ¡ì „ê³µ, ìŠ¬ë¦¼íƒ„íƒ„ \në‚´ìˆ­ì—…ì‹œ ì§ˆí€í•˜ê²Œ ì¿¨í•˜ê²Œ í•¨ í•˜ê³  \në¹ ì´ì¹˜ì‹¤ë¶„ ê° í†¡ì£¼ì„¸ìš”",
        "êµ¬ë˜ìŠ¬": "ë–¼êµ´ë–¼êµ´ë–¼êµ´"
}

    if character_name in easter_eggs:
        output += f"\n\n{easter_eggs[character_name]}"

    return output
```

```python
import time

# ì±„íŒ…ë°©ë³„ ë§ˆì§€ë§‰ í˜¸ì¶œ ì‹œê°„ ê¸°ë¡ìš© ë”•ì…”ë„ˆë¦¬
last_yugak_called_per_room = {}

def handle_yugak_command(chatroom_id):
    now = time.time()
    cooldown = 180  # 3ë¶„

    last_called = last_yugak_called_per_room.get(chatroom_id, 0)
    if now - last_called < cooldown:
        remaining = int(cooldown - (now - last_called))
        minutes = remaining // 60
        seconds = remaining % 60
        return f"â³ ì´ ëª…ë ¹ì–´ëŠ” í•´ë‹¹ ì¹´ì¹´ì˜¤í†¡ë°©ì—ì„œ {minutes}ë¶„ {seconds}ì´ˆ í›„ì— ë‹¤ì‹œ ì‚¬ìš© ê°€ëŠ¥í•©ë‹ˆë‹¤."

    # ì¿¨íƒ€ì„ ê°±ì‹ 
    last_yugak_called_per_room[chatroom_id] = now

    # ìœ ë¬¼ ê°ì¸ì„œ ê²€ìƒ‰ í•„í„° ì„¤ì •
    url = "https://developer-lostark.game.onstove.com/markets/items"
    payload = {
        'Sort': 'CURRENT_MIN_PRICE',
        'CategoryCode': 40000,
        'ItemGrade': 'ìœ ë¬¼',
        'SortCondition': 'DESC'
    }

    search_result = api_post(url, payload)
    if not search_result:
        return "âŒ ê°ì¸ì„œ ë°ì´í„°ë¥¼ ê°€ì ¸ì˜¬ ìˆ˜ ì—†ìŠµë‹ˆë‹¤."

    item_list = search_result.get("Items", [])
    if not item_list:
        return "âŒ ìœ ë¬¼ ê°ì¸ì„œê°€ ê²€ìƒ‰ë˜ì§€ ì•Šì•˜ìŠµë‹ˆë‹¤."

    item_prices = []
    for item in item_list:
        item_id = item.get("Id")
        detail_url = f"https://developer-lostark.game.onstove.com/markets/items/{item_id}"

        print(f"ğŸ§ª {item['Name']} (ID: {item_id})")
        detail_data_list = api_get(detail_url)

        if isinstance(detail_data_list, list) and detail_data_list:
            detail_data = detail_data_list[0]
            price = detail_data.get("CurrentMinPrice", 0)

            print(f"   ğŸ” CurrentMinPrice: {price}")

            if price > 0:
                item_prices.append({
                    "Name": item["Name"],
                    "Price": price
                })
        else:
            print("   âš ï¸ ìƒì„¸ ë°ì´í„° ì—†ìŒ")

    if not item_prices:
        return "âŒ ê°€ê²© ì •ë³´ë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤."

    sorted_items = sorted(item_prices, key=lambda x: x["Price"], reverse=True)[:10]

    result_lines = ["ğŸ’ ìœ ë¬¼ ê°ì¸ì„œ ìµœì €ê°€ TOP 10 (ë‚´ë¦¼ì°¨ìˆœ)"]
    for idx, item in enumerate(sorted_items, 1):
        result_lines.append(f"{idx}. {item['Name']}: {item['Price']:,} ê³¨ë“œ")

    result_lines.append("\nâ€» â³ ì´ ëª…ë ¹ì–´ëŠ” 3ë¶„ë§ˆë‹¤ í•œ ë²ˆë§Œ ì‚¬ìš© ê°€ëŠ¥í•©ë‹ˆë‹¤.")
    return "\n".join(result_lines)

# ë©”ì‹œì§€ í•¸ë“¤ëŸ¬ ì˜ˆì‹œ
def handle_kakao_message(msg, chatroom_id):
    if msg.strip() == "/ìœ ê°":
        return handle_yugak_command(chatroom_id)
    return "â“ ì§€ì›í•˜ì§€ ì•ŠëŠ” ëª…ë ¹ì–´ì…ë‹ˆë‹¤."
```
